```python
import numpy as np
import pandas as pd
from astropy import table
from pathlib import Path
import os
import matplotlib.pyplot as plt
from scipy import interpolate
import treecorr as tc
from astropy.io.misc.hdf5 import read_table_hdf5
from astropy import units as u
from astropy.cosmology import Planck18 as cosmo
from astropy.coordinates import SkyCoord

```

```python
# load in the elg correlation function data to compare

elg_clustering_data = np.load("/global/homes/y/yoki/roman/desi_like_samples/skysim_5000/data/desi_sv_data/desi_elg_ts_zenodo/elg_corr_data_z_0.8_1.1.npz") 
sep_08_11 = elg_clustering_data["sep"]
xis_08_11 = elg_clustering_data["xis"]
```

```python
# loop over all healpix and concat them together into one 
path_healpix_ids = '/global/homes/y/yoki/roman/desi_like_samples/skysim_5000/data/healpix_ids/id_nums_exclude_edges.npy'
hpix_ids = np.load(path_healpix_ids)
cat_list = []
rand_list = []


for hpix in hpix_ids:
    path_mock_elgs = f'/global/homes/y/yoki/roman/desi_like_samples/skysim_5000/data/mocks_per_healpix/mock_elg_cat_per_pixel/mock_elg_cat_hpix_{hpix}.parquet'
    mock_elg_cat_loop = pd.read_parquet(path=path_mock_elgs)
    cat_list.append(mock_elg_cat_loop)

mock_elg_cat = pd.concat(cat_list)

for hpix in hpix_ids:
    path_mock_elg_rands = f'/global/homes/y/yoki/roman/desi_like_samples/skysim_5000/data/rands_per_healpix/rands_elg_cat_per_pixel/rand_elg_cat_hpix_{hpix}.parquet'
    mock_elg_rand_loop = pd.read_parquet(path=path_mock_elg_rands)
    rand_list.append(mock_elg_rand_loop)

rand_elg_cat = pd.concat(rand_list)
```

```python
ra = mock_elg_cat['ra']
dec = mock_elg_cat['dec']
s = mock_elg_cat['s']
rp = mock_elg_cat['rp']

ra_rand = rand_elg_cat['ra']
dec_rand = rand_elg_cat['dec']
s_rand = rand_elg_cat['s']
rp_rand = rand_elg_cat['rp']
```

```python
# plt.scatter(mock_elg_cat['ra'], mock_elg_cat['dec'])
# plt.ylabel('DEC', fontsize=19 )
# plt.xlabel('RA', fontsize=19)
```

```python
# calculate the correlation using s

dg_to_r = np.pi / 180
MIN_SEP = 10
MAX_SEP = 200
NBINS = 25

tc_cat_s = tc.Catalog(ra=ra*dg_to_r, dec=dec*dg_to_r, r=s, ra_units='radians', dec_units='radians')
tc_rnd_s = tc.Catalog(ra=ra_rand*dg_to_r, dec=dec_rand*dg_to_r, r=s_rand, ra_units='radians', dec_units='radians')


dd_s = tc.NNCorrelation(min_sep=MIN_SEP, max_sep=MAX_SEP, nbins=NBINS)
dr_s = tc.NNCorrelation(min_sep=MIN_SEP, max_sep=MAX_SEP, nbins=NBINS)
rr_s = tc.NNCorrelation(min_sep=MIN_SEP, max_sep=MAX_SEP, nbins=NBINS)

dd_s.process(tc_cat_s, metric='Euclidean')
dr_s.process(tc_cat_s, tc_rnd_s, metric='Euclidean')
rr_s.process(tc_rnd_s, metric='Euclidean')
xi_s, _ = dd_s.calculateXi(rr=rr_s, dr=dr_s)


```

```python
s_h_scaled = cosmo.h * np.exp(dd_s.meanlogr)

fig, ax = plt.subplots()
ax.plot(s_h_scaled,  (s_h_scaled**(2))*xi_s, label='mock elg cat')
ax.plot(sep_08_11,  (sep_08_11**2)*xis_08_11[0], label='truth 0.8 < z < 1.1, l=0')
ax.set_xlabel(' s (Mpc/h)', fontsize=18)
ax.set_ylabel(r'$s^{2}\xi(s)$ $(\frac{Mpc}{h})^{2}$', fontsize=18)
plt.legend()
# ax.set_xscale('log')
# ax.set_yscale('log')
# ax.loglog()
#plt.savefig('/global/homes/y/yoki/roman/desi_like_samples/skysim_5000/figures/mock_elg_corr.png')
```
