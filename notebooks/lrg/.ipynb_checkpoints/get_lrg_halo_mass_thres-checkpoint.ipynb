```python
# load in the DESI LRG data
path_lrg_comoving_density = '/pscratch/sd/y/yoki/desi_like_data_diffsky/data/desi_sv_data/desi_lrg_ts_zenodo/fig_1_const_comoving_density.csv'
path_lrg_data = '/pscratch/sd/y/yoki/desi_like_data_diffsky/data/desi_sv_data/desi_lrg_ts_zenodo/fig_1_histograms.csv'

lrg_comoving_density = pd.read_csv(path_lrg_comoving_density, index_col=False)
desi_lrg_data = pd.read_csv(path_lrg_data, index_col=False)
zmin = desi_lrg_data['zmin']
zmax = desi_lrg_data['zmax']
n_lrg = desi_lrg_data['n_desi_lrg']
Z_GRID_POINTS = (len(n_lrg) * 32) + 1

z_bin_centers = (zmin + zmax ) / 2 
zgrid = np.linspace(np.min(zmin), np.max(zmax), Z_GRID_POINTS)
SELECTION_ZBIN_WIDTH = (np.max(zmax) - np.min(zmin)) / (Z_GRID_POINTS - 1)
```

```python
Z_GRID_POINTS
```

```python
repeat_n = int((Z_GRID_POINTS-1)/len(z_bin_centers))
new_zmin = np.linspace(np.min(zmin), np.max(zmax), Z_GRID_POINTS)[:-1]
new_zmax = np.linspace(np.min(zmin), np.max(zmax), Z_GRID_POINTS)[1:]
new_z_center = (new_zmax + new_zmin) / 2
new_n_lrg = np.repeat(n_lrg,repeat_n)/repeat_n

from scipy.interpolate import interp1d
interpf = interp1d(z_bin_centers, n_lrg, fill_value=0, bounds_error=False)
interp_n_lrg = interpf(new_z_center) / repeat_n
```

```python
print(np.sum(new_n_lrg), np.sum(n_lrg), np.sum(interp_n_lrg))
```

```python
values, edges = np.histogram(sim_cat['redshift_true'], bins=zgrid)
values_sim = values/TOTAL_SIM_AREA
z_frac = interp_n_lrg / values_sim


```

```python
z_frac
```

```python
z_frac = np.minimum(z_frac, np.ones(len(z_frac))*0.99)
```

```python
plt.hist(z_frac, bins=100)
plt.xlabel('z fraction')
plt.show()
```

```python
z_frac = np.array(z_frac)

mass_thres_list = []

for i in range(len(new_z_center)):
    print(i)
    this_zmin = new_zmin[i]
    this_zmax = new_zmax[i]

    this_cat = sim_cat[np.logical_and(sim_cat['redshift_true']>this_zmin, sim_cat['redshift_true']<this_zmax)]

    if len(this_cat) < 100:
        print(f"Empty bin: zmin={this_zmin}, zmax={this_zmax}")
        this_mass_thres = 10**30
    else:
        print(len(this_cat))
        this_mass_thres = np.percentile(a = this_cat['log_halo_mass'], q = 100-z_frac[i]*100)
        
        
        
    mass_thres_list.append(this_mass_thres)

mass_thres = np.array(mass_thres_list)

```

```python
path_mass_thres = '/pscratch/sd/y/yoki/desi_like_data_diffsky/data/selection_thresholds/lrg/halo_mass_thres.npy'
np.save(path_mass_thres, mass_thres)
```

```python
plt.plot(new_z_center, (mass_thres), c='C0')
plt.xlabel('redshift')
plt.ylabel('Log10(Halo Mass) threshold')
# plt.plot(z_bin_centers, (smooth_thres), c='C1')
# plt.plot(z_fine_grid, (sfr_fine_grid), ls='--' , c='C2')
# plt.yscale('log')
plt.xlim(0,2)
plt.ylim(9,20)
```
